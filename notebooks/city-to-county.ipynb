```python
cities = pd.read_csv('../data/fatal-police-shootings-data-COMPLETE.csv', engine='python')
cities = cities[['city', 'state']]
cities = cities.drop_duplicates().reset_index(drop=True) # keep only unique locations
print(len(cities))
cities.head()
```

commented out to avoid overrunning the API

```python
location_list = []
city_list = []
state_list = []

for data in cities.iterrows():
    city = data[1]['city']
    state = data[1]['state']
    query = f"{city}, {state}"
    try:
        print(f"{data[0]}: {query}")
        location = google_maps.geocode(query)
        location_list.append(location)
        city_list.append(city)
        state_list.append(state)
    except:
        print(f"{query} NOT FOUND")
```

save all info to json

```python
with open('../data/location_data.json', 'w') as fout:
    json.dump(location_list, fout)
```

```python
county_list = []

for location in location_list:
    try:
        county = find_county(location)
        county_list.append(county)
    except:
        print(f"{location[0]['formatted_address']} county not found")
        county_list.append("")
```
